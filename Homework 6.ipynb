```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Random Plots

# generating lists
list1 = np.random.randint(0,101,40)
list2 = np.random.randint(0,101,40)
list3 = np.random.randint(0,101,40)

# Values for top plot
x1 = np.linspace(0,101,40)
y1 = list1

x2 = np.linspace(0,101,40)
y2 = list2

# Values for bottom plot
x3 = np.linspace(0,101,40)
y3 = list3

# Graphing
fig, axs = plt.subplots(2,1)

plt.title('graph')
plt.xlabel('position')
plt.ylabel('plots')

axs[0].errorbar(x1,y1, color = 'orange', linewidth = 10)
axs[0].errorbar(x2,y2, color ='r', linestyle = 'dashed')

axs[1].scatter(x3,y3, color = 'm', marker = 'D')

plt.show()
```

```python
#2. van der Waals Gas

# Ideal gas law: (P + (an^2)/V^2)(V-nb) = nRT
# # Where P is pressure, V is volume, T is temperature, n is number of moles, R is the ideal gas constant and
# a, b are parameters.

# a = 16.02                # parameter
# b = 0.1124               # parameter
# R = 0.083144             # Ideal Gas Constant
# n = 1                    # number of moles (1 mole of acetone)
# P = np.linspace(0,11)      # Pressure
# V = np.linspace(10,31)     # Volume

P, V = np.meshgrid(np.linspace(0,11,50), np.linspace(10,31,50))

# Equation for temperature derived from the Ideal Gas Law
T = (P + 16.02/V**2)*(V - 0.1124)/(0.083144)

# graphing the data
plt.figure()
plt.imshow(T, cmap = 'rainbow', vmin = 0, vmax = 3500)
plt.xlabel('Pressure (Bar)')
plt.ylabel("Volume (Liters)")
plt.colorbar()
plt.title("Temperature for One Mole of Acetone")
plt.show()
```

```python
# 3. Monte Carlo 

# generating values
n = 10**4
x = 1-np.random.rand(int(n))
y = 1-np.random.rand(int(n))
x_in, y_in = x[x**2 + y**2 <= 1], y[x**2 + y**2 <= 1]
x_out, y_out = x[x**2 + y**2 > 1], y[x**2 + y**2 > 1]

# Graphing the plot
fig, axs = plt.subplots(1)
axs.scatter(x_in, y_in, c = 'r')
axs.scatter(x_out, y_out, c = 'b')
plt.title('Monte Carlo')
plt.show

# Printing pi
n_in = 0
for i in range(0,n):
    x = np.random.rand()
    y = np.random.rand()
    if np.sqrt(x**2 + y**2) <= 1:
        n_in += 1
pi = (n_in/n)*4
print(pi)
```
